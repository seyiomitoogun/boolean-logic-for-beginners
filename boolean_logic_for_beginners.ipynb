Boolean Operators Explained in True Terms:
## Understanding Boolean Operators in Everyday Life

Boolean operators help us make decisions by comparing things.  
They answer questions like:
- Is this equal to that?
- Is this bigger or smaller?
- Do both conditions need to be true?
- Is one condition enough?

Letâ€™s explore each operator with a real-life example.


```python
# Equal To (==)
# Use when you want to check if two things are the same.
# Example: Is today Monday?
today = "Monday"
print(today == "Monday")  # True

# Not Equal To (!=)
# Use when you want to check if two things are different.
# Example: Is today NOT Sunday?
print(today != "Sunday")  # True

# Greater Than (>)
# Use when you want to check if one number is bigger than another.
# Example: Is my age greater than 60?
age = 65
print(age > 60)  # True

# Less Than (<)
# Use when you want to check if one number is smaller than another.
# Example: Is my shoe size less than 10?
shoe_size = 8
print(shoe_size < 10)  # True

# Greater Than or Equal To (>=)
# Use when you want to include the number you're comparing to.
# Example: Do I qualify for the senior discount (60 and above)?
print(age >= 60)  # True

# Less Than or Equal To (<=)
# Use when you want to include the lower limit.
# Example: Is my blood pressure within safe range (120 or less)?
bp = 118
print(bp <= 120)  # True
```

Summary: Youâ€™ve just learned how to compare values in Python using everyday logic. Each operator is like a question we ask to make wise decisions.  



Step-by-Step: Categorizing Transactions with Python

```python
# Sample list of transactions
transactions = [
    {"date": "2025-11-01", "description": "Walmart", "amount": 45.20},
    {"date": "2025-11-02", "description": "Shell Gas", "amount": 30.00},
    {"date": "2025-11-03", "description": "Office Depot", "amount": 120.50},
    {"date": "2025-11-04", "description": "Panera Bread", "amount": 18.75},
]

# Categorize each transaction
for tx in transactions:
    desc = tx["description"].lower()

    if "walmart" in desc or "panera" in desc:
        category = "Supplies & Meals"
    elif "shell" in desc:
        category = "Transportation"
    elif "office" in desc:
        category = "Office Expenses"
    else:
        category = "Uncategorized"

    print(f"{tx['date']} | {tx['description']} | ${tx['amount']} â†’ {category}")
```

Uses if, elif, and else to check keywords in the transaction description.

Assigns a category based on whatâ€™s found.

Prints a clean summary line for each transaction.
